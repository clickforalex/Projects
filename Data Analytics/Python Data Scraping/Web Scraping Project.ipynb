```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Singapore'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find_all('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[2]
```

```python
company_title = table.find_all('th')
```

```python
print(company_title)
```

```python
singapore_table_titles = [title.text.strip() for title in company_title]
print(singapore_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = singapore_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_data
```
